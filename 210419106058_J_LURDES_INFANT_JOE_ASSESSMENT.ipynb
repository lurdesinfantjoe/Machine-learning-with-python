### Q2. Build time series model and check the performance of the model using RMSE

### Linear Model

```python
df_mumbai = market_df.loc[market_df.city == "MUMBAI"].copy()
df_mumbai.head()
```

```python
df_mumbai = df_mumbai.drop(["market","month","year","priceMin","priceMax","state","city"], axis=1)
df_mumbai.head()
```

```python
df_mumbai["log_priceMod"] = np.log(df_mumbai.priceMod)
```

```python
df_mumbai_mean = df_mumbai.log_priceMod.mean()
df_mumbai["mean_price"] = np.exp(df_mumbai_mean)
```

```python
df_mumbai["timeindex"] = df_mumbai.date - df_mumbai.date.min()
df_mumbai.head()
```

```python
df_mumbai["timeindex"] = df_mumbai["timeindex"]/np.timedelta64(1,"M")
df_mumbai.head()
```

```python
df_mumbai["timeindex"] = df_mumbai["timeindex"].round(0).astype(int)
df_mumbai.tail()
```

```python
linear_model = smf.ols('log_priceMod ~ timeindex', data = df_mumbai).fit()
linear_model.summary()
```

```python
linear_model_pred = linear_model.predict()
linear_model_pred
```

```python
df_mumbai["linear_price"] = np.exp(linear_model_pred)
df_mumbai.head()
```

```python
model_linear_forecast_manual = 0.0093 * 146 + 6.1121
model_linear_forecast_manual
```

```python
def RMSE(actual, predicted):
    mse = (actual - predicted)**2
    rmse = np.sqrt(mse.sum()/mse.count())
    return rmse
```

```python
linear_model_RMSE = RMSE(df_mumbai.priceMod, df_mumbai.linear_price)
linear_model_RMSE
```

```python
np.exp(model_linear_forecast_manual)
```

```python
Result_df = pd.DataFrame(columns =["Model","Forcast","RMSE"])
Result_df.loc[0,"Model"] = "Linear Model"
Result_df.loc[0,"Forcast"] = np.exp(model_linear_forecast_manual)
Result_df.loc[0,"RMSE"] = linear_model_RMSE
Result_df
```

### Thus we have build a time series model and checked the performance of the model using RMSE

### Q3. Plot ACF and PACF plots 

```python
df_mumbai["shift_log_priceMod"] = df_mumbai.log_priceMod.shift()
df_mumbai["log_priceMod_diff"] = df_mumbai.log_priceMod - df_mumbai.shift_log_priceMod
```

```python
ts = df_mumbai.log_priceMod
ts_diff = df_mumbai.log_priceMod_diff
ts_diff.dropna(inplace = True)
```

```python
from statsmodels.tsa.stattools import acf, pacf
lag_acf = acf(ts_diff, nlags=20)
lag_acf
```

### ACF Plot

```python
plt.figure(figsize=(10,10))
ACF = pd.Series(lag_acf)
ACF.plot(kind="bar")
```

### PACF plot

```python
lag_pacf = pacf(ts_diff, nlags=20, method='ols')
plt.figure(figsize=(10,10))
PACF = pd.Series(lag_pacf)
PACF.plot(kind="bar")
```

### Thus we have plotted the ACF and PACF plots
